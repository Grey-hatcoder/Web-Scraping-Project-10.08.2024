```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from pprint import pprint
import pandas as pd
```

```python
def book_url_to_dict(a: str) -> dict:
    book_page = requests.get(a)
    book_soup = BeautifulSoup(book_page.content, parser = "html.parser")
    book_title = book_soup.find(name = 'h1').get_text()
    book_tr = book_soup.find_all(name = 'tr')
    book_dict = {}
    book_dict['Title'] = book_title
    for i in book_tr:
        book_dict[i.findChild(name = 'th').getText()] = i.findChild(name = 'td').getText()
    return book_dict
```

```python
def Page_to_book_urls(page: str) -> list:
    HP = requests.get(page)
    soup = BeautifulSoup(HP.content, parser = "html.parser")
    books_comp = soup.find_all(name = 'li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
    return_list = []
    for i in books_comp:
        book_anc = i.findChild(name = "a")
        Extended_url = book_anc.get("href")
        book_url = urljoin(page,Extended_url)
        return_list.append(book_url)
    return return_list
```

```python
data = []
p = 1
page = f'https://books.toscrape.com/catalogue/page-{p}.html'
while(requests.get(page).status_code==200):
    book_urls = Page_to_book_urls(page)
    for i in book_urls:
        data.append(book_url_to_dict(i))
        #print(book_url_to_dict(i)['Title'])
    p+=1
    print(p)
    page = f'https://books.toscrape.com/catalogue/page-{p}.html'
#print(data)
```
